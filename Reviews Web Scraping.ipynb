```python
url = "https://www.amazon.in/product-reviews/B08D9NDZ1Y/ref=acr_dp_hist_1?ie=UTF8&filterByStar=one_star&reviewerType=all_reviews#reviews-filter-bar"
```

```python
driver.get(url)
```

```python
UR=[]
for i in range(0,150):
    urls=driver.find_elements_by_xpath("//a[@class='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold']")
    for i in urls:
        UR.append(i.get_attribute('href'))# getting the urls
        
    time.sleep(3)
    
    url21=driver.find_element_by_xpath("//li[@class='a-last']/a").get_attribute("href")
    time.sleep(2)
    driver.get(url21) 
```

```python
review=[]
```

```python
for url in UR: #loop for every review in the list
        driver.get(url)
        try:
            rev=driver.find_element_by_xpath("//span[@class='a-size-base review-text review-text-content']/span") # locating the ratings
            review.append(rev.text)
        
        except NoSuchElementException as e:
            review.append('No review')
```

```python
ratings=[]
```

```python
for url in UR: #loop for every review in the list
        driver.get(url)
        try:
            rating=driver.find_element_by_xpath("//a[@class='a-link-normal']") # locating the ratings
            ratings.append(rating.get_attribute('title'))
        
        except NoSuchElementException as e:
            ratings.append('No ratings')
```

```python
df=pd.DataFrame({'Review':review,
                     'Ratings':ratings
                     })
```

```python
print(df)
```

```python
df.to_excel('Ratings_prediction3.xlsx', index = False)
```
